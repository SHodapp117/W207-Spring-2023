You will likely have to upload the file `players_20.csv` to the workspace yourself.

```
df = pd.read_csv("/players_20.csv")
df.head()
```

The Pandas documentation [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#applying-multiple-functions-at-once) should be a useful reference.

```
# Grouping the data by nationality and calculating count and average age
grouped = df.groupby('nationality').agg(
    player_count=pd.NamedAgg(column='short_name', aggfunc='count'),
    average_age=pd.NamedAgg(column='age', aggfunc='mean')
).reset_index()
```

```
import altair as alt

# Create a scatter plot with a title and custom scale domains, with x and y axes flipped
scatter_plot = alt.Chart(grouped).mark_circle().encode(
    y=alt.Y('player_count:Q', scale=alt.Scale(domain=(0, 1700))),  # Custom y-axis scale domain
    x=alt.X('average_age:Q', scale=alt.Scale(domain=(20, 40))),   # Custom x-axis scale domain
    color='nationality:N',  # Color points by nationality
    tooltip=['nationality', 'player_count', 'average_age']  # Show tooltips on hover
).properties(
    width=600,  # Width of the plot
    height=400,  # Height of the plot
    title='Player Counts and Average Ages by Nationality'  # Set the title
).configure_title(
    fontSize=16,  # Font size of the title
    anchor='start',  # Title anchor position
    offset=10  # Title offset from the top of the plot
)

# Show the scatter plot
scatter_plot
```
